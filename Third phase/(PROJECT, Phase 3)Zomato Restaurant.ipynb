```python
import pandas as pd

# Load datasets
zomato_data = pd.read_csv("https://github.com/dsrscientist/dataset4/raw/main/zomato.csv", encoding='ISO-8859-1')
zomato_data
```

```python
country_code_data = pd.read_excel("https://github.com/dsrscientist/dataset4/raw/main/Country-Code.xlsx")
country_code_data
```

```python
''' Data Cleaning'''
# Handle missing values
zomato_data.dropna(inplace=True)
```

```python
# Merge datasets based on the 'Country Code' column
df = pd.merge(zomato_data, country_code_data, on='Country Code', how='left')
```

```python
df
```

```python
# Display basic information about the datasets
print(df.info())
```

```python
df.shape # to check the dimension of the data set (rows, columns)
```

```python
# Feature extraction
features = df[['Cuisines', 'Has Table booking', 'Has Online delivery', 'Aggregate rating']]

# Target variables
target_avg_cost = df['Average Cost for two']
target_price_range = df['Price range']

```

```python
'''Label Encoding or One-Hot Encoding for categorical variables'''
# Using One-Hot Encoding for 'Cuisines' and binary encoding for 'Has Table booking' and 'Has Online delivery'

# One-Hot Encoding for 'Cuisines'
features = pd.get_dummies(features, columns=['Cuisines'], drop_first=True)

# Binary Encoding for 'Has Table booking' and 'Has Online delivery'
features['Has Table booking'] = features['Has Table booking'].map({'Yes': 1, 'No': 0})
features['Has Online delivery'] = features['Has Online delivery'].map({'Yes': 1, 'No': 0})
```

```python
'''Model Building'''

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score

# Split the data into training and testing sets
X_train, X_test, y_train_avg_cost, y_test_avg_cost = train_test_split(features, target_avg_cost, test_size=0.2, random_state=42)
X_train_price_range, X_test_price_range, y_train_price_range, y_test_price_range = train_test_split(features, target_price_range, test_size=0.2, random_state=42)


```

```python
# Linear Regression for predicting average cost
regression_model = LinearRegression()
regression_model.fit(X_train, y_train_avg_cost)
predictions_avg_cost = regression_model.predict(X_test)
```

```python
# Logistic Regression for predicting price range
classification_model = LogisticRegression(max_iter=1000)
classification_model.fit(X_train_price_range, y_train_price_range)
predictions_price_range = classification_model.predict(X_test_price_range)
```

```python
'''Model Evaluation'''

# Evaluate regression model
mse = mean_squared_error(y_test_avg_cost, predictions_avg_cost)
print(f"Mean Squared Error (MSE) for Average Cost prediction: {mse}")

# Evaluate classification model
accuracy = accuracy_score(y_test_price_range, predictions_price_range)
print(f"Accuracy for Price Range prediction: {accuracy}")
```

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

# Random Forest for predicting average cost
rf_regressor = RandomForestRegressor()
rf_regressor.fit(X_train, y_train_avg_cost)
predictions_avg_cost_rf = rf_regressor.predict(X_test)

# Random Forest for predicting price range
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train_price_range, y_train_price_range)
predictions_price_range_rf = rf_classifier.predict(X_test_price_range)
```

```python
from sklearn.metrics import mean_squared_error

# Evaluate Random Forest regression model
mse_rf_avg_cost = mean_squared_error(y_test_avg_cost, predictions_avg_cost_rf)
print(f"Mean Squared Error (MSE) for Random Forest Average Cost prediction: {mse_rf_avg_cost}")
```

```python
from sklearn.metrics import accuracy_score

# Evaluate Random Forest classification model
accuracy_rf_price_range = accuracy_score(y_test_price_range, predictions_price_range_rf)
print(f"Accuracy for Random Forest Price Range prediction: {accuracy_rf_price_range}")
```
